```python
X = vendor_df[['Dollars']]
```

```python
Y = vendor_df[['Freight']]
```

```python
vendor_df.describe().round()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , Y_train, Y_test =train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import *
```

```python
model1= LinearRegression()
model1.fit(X_train , Y_train)

model2 = DecisionTreeRegressor(max_depth= 4,random_state=42)
model2.fit(X_train , Y_train)

model3 = RandomForestRegressor(max_depth=4 , random_state=42)
model3.fit(X_train , Y_train)
```

```python
def evaluate_model(model , X_test , Y_test , model_name):
    preds = model.predict(X_test)

    mae= mean_absolute_error(Y_test, preds)
    rmse = mean_squared_error(Y_test , preds, squared = False)
    r2 = r2_score(Y_test , preds)*100

    print(f"\n{model_name} Performence:")
    print(f"MAE : {mae:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"R2 score : {r2 :.2f}")
    
```

```python
evaluate_model(model1 , X_test, Y_test , 'LinearRegression')
```

```python
evaluate_model(model2 , X_test, Y_test , 'DecisonTree')
```

```python
evaluate_model(model3 , X_test, Y_test , 'RandomForest')
```
